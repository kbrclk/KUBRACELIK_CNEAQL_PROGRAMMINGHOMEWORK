Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House:
  def __init__(self, area, rooms, bathrooms, floors, garage, garden, pool, fireplace=False,elevator=False):
    self.area = area
    self.rooms = rooms
    self.bathrooms = bathrooms
    self.floors = floors
    self.garage = garage
    self.garden = garden
    self.pool = pool
    self.fireplace = fireplace
    self.elevator = elevator
  def price(self):
    price = (self.area * self.floors *  1500 + self.rooms * 500 + self.bathrooms * 500 + self.garage * 10000 +
             self.garden * 15000 + self.pool * 12000 + self.fireplace * 9000 + self.elevator * 10000)
    return price
  def __str__(self):
    return f"""This house futures:
    Area: {self.area}
    Rooms: {self.rooms}
    Bathrooms: {self.bathrooms}
    Floors: {self.floors}
    Garage: {self.garage}
    Garden: {self.garden}
    Pool: {self.pool}
    Fireplace: {self.fireplace}
    Elevator: {self.elevator}
    This house price is {self.price()}$"""
```

*   name

```python
class Person:
  def __init__(self, name, age, married, children, cars, money):
    self.name = name
    self.age = age
    self.married = married
    self.children = children
    self.cars = cars
    self.money = money
    if self.married == True:
      self.family_members = self.children + 2
    else:
      self.family_members = self.children + 1

  def __str__(self):
    return f"""This person's futures:
    Name: {self.name}
    Age: {self.age}
    Married: {self.married}
    Children: {self.children}
    Cars: {self.cars}
    Money: {self.money}$
    Family Members: {self.family_members}"""
```

*   whether the person can buy the house,
*   whether the person's family can live in that house (enough room),
*   whether the house has enough space (garages) for the cars in the family

(5 points)


```python
def can_buy_house(person, house):
    if person.money >= house.price():
        return f"{person.name} can buy {house.price()}$ house."
    else:
        return f"{person.name} cannot buy this house, needs {house.price()}$."
```

```python
Person.can_buy_house = can_buy_house
```

```python
def can_family_live(person, house):
    if house.rooms >= person.family_members:
        return f"Yes! The house has {house.rooms} rooms."
    else:
        return f"No! The house only has {house.rooms} rooms."
```

```python
Person.can_family_live = can_family_live
```

```python
def has_enough_garage(person, house):
    if house.garage >= person.cars:
        return f"Yes! The house has {house.garage} garage."
    else:
        return f"No! The house has not {person.cars} garage."
```

```python
Person.has_enough_garage = has_enough_garage
```

 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

```python
class Animal:
  """Which methods belong to parent: Dinosaurs, chickens, and ostriches share evolutionary links.
  So they share basic characteristics such as egg-laying and some limbs."""

  def __init__(self, name, legs, wings, lays_eggs=True):
    self.name = name
    self.legs = legs
    self.wings = wings
    self.lays_eggs = lays_eggs
  def make_sound(self):
    return "Some animal sound"
  def fly(self):
    return "This animal's flying ability is unknown."
  def __str__(self):
    return f"""This animal's futures:
    Name: {self.name}
    Legs: {self.legs}
    Wings: {self.wings}
    Lays eggs: {self.lays_eggs}"""

class Dinosour(Animal):
  def __init__(self, name, legs, wings, exist=False):
    super().__init__(name, legs, wings)
    self.exist = exist

  """- Which should be overridden: Subclasses override the 'make_sound' and 'fly' methods to exhibit different behaviors."""

  def make_sound(self):
    return "Rawr"
  def fly(self):
    if self.wings:
      return "This dinosaur can fly."
    else:
      return "This dinosaur cannot fly."

  """- Which attributes are unique: Each subclass extends state and behavior (e.g., extinction, incubation_period and edibility)."""

  def extinction(self):
        if self.exist:
          return f"Wow, {self.name} is still alive!"
        return f"Sadly, {self.name} is extinct."


class Chicken(Animal):
  def __init__(self, name, incubation):
    super().__init__(name, legs=2, wings=True)
    self.incubation = incubation
  def make_sound(self):
    return "Cluck"
  def fly(self):
    return f"Technically yes, but {self.name} can only fly for a few seconds over short distances."
  def incubation_period(self):
    if self.incubation==21:
      return "There is a new chick in the world."
    return "Sadly, incubation is unsuccessful."

class Ostrich(Animal):
  def __init__(self, name, edible=True):
    super().__init__(name, legs=2, wings=True)
    self.edible = edible
  def make_sound(self):
    return "Screech"
  def fly(self):
    return f"No! {self.name} has wings, but it is a flightless bird."
  def edibility(self):
    if self.edible:
        return f"{self.name} meat is healthier than beef."
    else:
        return f"WARNING: {self.name} is an endangered species. Please do not consume!"
```

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
class Shape:
  def __init__(self,name):
    self.name = name
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    super().__init__(name = "Circle")
    self.radius = radius
  def area(self):
    return 3.14*self.radius**2

class Square(Shape):
  def __init__(self, side):
    super().__init__(name = "Square")
    self.side = side
  def area(self):
    return self.side**2

class Rectangle(Shape):
  def __init__(self, length, width):
    super().__init__(name = "Rectangle")
    self.length = length
    self.width = width
    if self.length == self.width:
            self.name = "Square (Created from Rectangle)"
  def area(self):
    return self.length*self.width
```

```python
c1 = Circle(3)
```

```python
s1 = Square(5)
```

```python
r1 = Rectangle(3,5)
```

```python
shapes = [c1,s1,r1]
for shape in shapes:
  print(f"{shape.name} area: {shape.area()}")
```

Count character frequency

```python
text = "banana"
```

Expected output: {'b': 1, 'a': 3, 'n': 2}

```python
chr = {}
for i in text:
  if i not in chr:
    chr[i] = 1
  else:
    chr[i] += 1
print(chr)
```
